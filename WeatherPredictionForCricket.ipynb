```python
#NaiveBayes Project
import pandas as pd 
from sklearn.preprocessing import LabelEncoder #this module is used to convert strings to numbers
from sklearn.naive_bayes import GaussianNB
```

```python
#Reading the csv files
df = pd.read_csv('weatherDataset.csv')
df
```

```python
#Encoding the string to numericals
Numerics = LabelEncoder()
```

```python
#Dropping the play column in the df and store it in the target column
inputs = df.drop('Play', axis='columns')
print("target")
target = df['Play']
target
```

```python
#Creating dataframes with numeric values for their string counterpart
inputs['OutlookNum'] = Numerics.fit_transform(inputs['Outlook'])
inputs['TempNum'] = Numerics.fit_transform(inputs['Temp'])
inputs['HumidityNum'] = Numerics.fit_transform(inputs['Humidity'])
inputs['WindNum'] = Numerics.fit_transform(inputs['Wind'])
inputs
```

```python
#Dropping the string values from the dataframe
inputsNum = inputs.drop(['Outlook', 'Temp', 'Humidity', 'Wind'], axis='columns')
inputsNum
```

```python
#Applying the Gaussian NaiveBayes by creating a classifer
Classifier = GaussianNB()
Classifier.fit(inputsNum, target)
```

```python
#Checking its Accuracy.. 90.9%
Classifier.score(inputsNum, target)
```

```python
#Prediction 
Result = Classifier.predict([[1, 1, 0, 0]])
Result[0]
```

```python
import joblib
joblib.dump(Classifier, 'weatherDataModel.pkl')
```
